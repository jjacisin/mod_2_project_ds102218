```python
def create_summary_df():
    path = '/Users/jim_jack/Documents/mod_2_project/mod_2_project_ds102218/results_csvs'
    allFiles = glob.glob(path + "/*.csv")
    df_summary = []
    for data in allFiles:
        df_dict = {}
        local_df = pd.read_csv(data)
        df_dict['Race'] = file_name_finder(data)
        if local_df.iloc[0]["County"] == "ALL COUNTIES":
#             print('standard - '+str(df_dict))
            df_dict['REP'] = local_df[local_df["County"]=="ALL COUNTIES"]["REP"][0]
            df_dict['Total Votes'] = local_df[local_df["County"]=="ALL COUNTIES"]["Total Votes"][0]
        else:
#             print("double check - "+str(df_dict))
            df_dict['REP'] = local_df.iloc[0]["REP"]
            df_dict['Total Votes'] = local_df.iloc[0]["Total Votes"]
        
        df_summary.append(df_dict)
    return pd.DataFrame(df_summary)
```

```python
df_draft = create_summary_df()
df_draft
```

```python
def add_leading_zero(row):
    if len(row) == 4:
        return row[0:3] +str(0) + row[-1] 
    else:
        return row
```

```python
df_draft["Race"] = df_draft["Race"].apply(add_leading_zero)
df_draft
# df.loc[df['First Season'] > 1990, 'First Season'] = 1
```

```python
df_cleaned = df_draft.set_index("Race")
df_sorted = df_cleaned.sort_index()
df_sorted
```

```python
def prop_r_vote(row):
    return int(row['REP'])/int(row['Total Votes'])
df_sorted['prop_r_vote'] = df_sorted.apply(lambda row: prop_r_vote(row),axis=1)
```
